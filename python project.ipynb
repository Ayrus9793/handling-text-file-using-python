```python
import pandas as pd
```

```python
data = pd.read_csv('desktop\conv.txt',delimiter= ':',header =None)
```

```python
data.columns=['speaker','speech']
```

```python
data.to_csv('desktop\conv.csv',index=None)
```

# Different speaker in the conv.txt


```python
unique = []

for speaker_name in data.speaker:
    if speaker_name not in unique:
        unique.append(speaker_name)
        
unique.sort()

print('\nNumber of unique characters in text file ',len(unique))
```

# Code to Store unique words of different Charatcters

```python

data_file  = open("desktop\conv.txt",'r')
lines = data_file.readlines()
data_file.close()

write_file = open("desktop\Arya.txt",'w')

for line in lines:
    if "ARYA"  in line:
        write_file.write(line)
        write_file.write("\n")
write_file.close()
```

```python
read_file = open("desktop\Arya.txt",'r')
lines = read_file.read()
lines = lines.lower()
data=lines.split()

data = [line.strip(',.?!:¦;()[]') for line in data]
data = [line.replace('\'sâ€™',"") for line in data]


unique2 = []
for line in data:
    if line not in unique2:
        unique2.append(line)
unique2.sort()


read_file.close()

rewrite_file = open("desktop\Arya.txt",'w')
for x in unique2:
    rewrite_file.write(x)
    rewrite_file.write("\n")
    
rewrite_file.close()

```
